# Simulating Smallpox Outbreak In New York City#

## Question ##
How would different responses to a smallpox outbreak in NY City impact the population?

## Model ##
In this model, we simulate all the different stages of smallpox including death and recovery. In the US today, virtually no one is vaccinated against smallpox since it has been irradicated. However, stocks of smallpox still exist in the US and Russia. If the smallpox disease was released in a concentrated urban area such as New York City in the context of bioterrorism, we can see how different responses to an outbreak would affect the outcome. 

```python
# Configure Jupyter so figures appear in the notebook
%matplotlib inline

# Configure Jupyter to display the assigned value after an assignment
%config InteractiveShell.ast_node_interactivity='last_expr_or_assign'

# import functions from the modsim library
from modsim import *
```

In this state, we create our seven stocks which represents all the stages of smallpox, susceptible, dead, and immune populations. 

```python
def make_state(S, U, I, E, R, D, N):
    """Makes a state.
       S: Susceptible
       U: Incubation
       I: Initial Symptoms
       E: Early Rash
       R: Late Rash to Scabs Falling Off
       D: Death
       N: Immune
    Returns a state."""
    
    state = State(S=S, U=U, I=I, E=E, R=R, D=D, N=N)
    return state
```

In the system, we have three transmission rates which correspond to the three infectious stocks (initial symptoms, early rash, and rash/scabs).

```python
def make_system(tr1, tr2, tr3, dt, t0, tend):
    """Makes a system.
    
    tr1: Transmission rate for initial symptoms. (Least contagious stage)
    tr2: Transmission rate for early rash. (Most contagious stage)
    tr3: Transmission rate for rash/scabs. (Relatively contagious stage)
    dt: Time step.
    t0: Start time.
    tend: End time.
    
    Returns a system."""
    
    system = System(tr1=tr1,
                   tr2=tr2,
                   tr3=tr3,
                   dt=dt,
                   t0=t0,
                   tend=tend)
    return system
```

We are calculating beta in each time step because the contact rate changes with the number of infectious and susceptible people. We start with calculating the proportion of the susceptible population each person in one of the infectious stages comes into contact with. Then, we calculate the contact rate for each infectious stage by multiplying the transmission rate with the proportion of susceptible people interacted with. After, we calculate the "weight" or proportion each stage makes up of the total infected population. Finally, we calculate the weighted average by multiplying this "weight" with the corresponding contact rates to find the final beta. 

As a safety measure, we make sure that the function doesn't keep trying to calculate beta once the number of susceptible people reaches zero because there is no one left to infect.

```python
def calc_beta(state, system, num_init, num_early_rash, num_rash):
    """Makes a system.
    
    state: State object
    system: System object
    num_init: Number of people met during initial symptoms. (Most people contacted)
    num_early_rash: Number of people met during early rash. (Few people contacted)
    num_rash: Number of people met during rash/scabs. (Few people contacted)
    prop_init:Proportion of susceptible pop contacted per person during intial stage.
    prop_early_rash:Proportion of susceptible pop contacted per person during early rash stage.
    prop_rash:Proportion of susceptible pop contacted per person during rash/scab stage.
    c_init: Contact rate for initial symptoms. (Least contagious stage)
    c_early_rash: Contact rate for early rash. (Most contagious stage)
    c_rash: Contact rate for rash/scabs. (Relatively contagious stage)
    
    Returns a system."""
    #making sure never divides by zero when calculating beta
    if state.S <= 0:
        beta = 0
    else:
        #calculating proportions
        prop_init = num_init/state.S            
        prop_early_rash = num_early_rash/state.S
        prop_rash = num_rash/state.S
    
        #calculating contact rates
        c_init = prop_init*tr1      
        c_early_rash = prop_early_rash*tr2                    
        c_rash = prop_rash*tr3 
    
        #calculating proportion of infected people
        prop_I = state.I/(state.I+state.E+state.R)
        prop_E = state.E/(state.I+state.E+state.R)
        prop_R = state.R/(state.I+state.E+state.R)
    
        #calculating weighted average
        beta = ((c_init*prop_I)+(c_early_rash*prop_E)+(c_rash*prop_R))
 
    
    return beta
```

This function starts by calculating beta for each time step, so that the most recent stock values are affecting the total contact rate. We calculate i, the total infected population. This does not include people in the incubation stock because they are not infectious, although they are infected with smallpox. We calculate the amount of people moving between each stock so that we know that int does not round differently between the differential equations.In u_to_i, we divide the stock U by 12 because the mean amount of days to move from incubation to initial symptoms is 12 days. Thus also applies in i_to_e, e_to_r, and r_to_rest where the average time in each stock is 3, 4, and 16 days. 

We assume that 70% of people will survive smallpox and therefore become immune and that 30% will not survive the disease. We have an if-else statement that prevents the number of susceptible people from becoming negative. Lastly, we update each of our stocks.

```python
def update_function(state, t, system, calc_beta, num_init, num_early_rash, num_rash, immunize):
    """Update the smallpox model.
    
    state: State object
    t: Time
    system: System object
    calc_beta: Function that updates beta.
    num_init: Number of people met during initial symptoms. (Most people contacted)
    num_early_rash: Number of people met during early rash. (Few people contacted)
    num_rash: Number of people met during rash/scabs. (Few people contacted)
    
    Returns a state."""
    
    #calculates beta
    beta = calc_beta(state, system, num_init, num_early_rash, num_rash)
    
    #total contagious people
    i = state.I+state.E+state.R
    
    #making sure the int doesn't round differently
    s_to_u = int(beta * state.S * i)
    immune = int(immunize*state.S)
    u_to_i = int(state.U/12)
    i_to_e = int(state.I/3)
    e_to_r = int(state.E/4)
    r_to_rest = int(state.R/16)
    
    #change in S for each time step
    dsdt = -s_to_u - immune
    #change in U, 12 days to move to I
    dudt = s_to_u - u_to_i
    #change in I, 3 days to move to E
    didt = u_to_i - i_to_e
    #change in E, 4 days to move to R
    dedt = i_to_e - e_to_r
    #change in R
    drdt = e_to_r - r_to_rest
    #change in N, after 16 days 70% recover
    dndt = int(r_to_rest * 0.7) + immune
    #change in D, after 16 days 30% die
    dddt = int(r_to_rest * 0.3)
    
 
    #updating the stocks with Euler's method
    #making sure S never goes below zero
    if (state.S + dsdt) <= 0:
        state.S = 0
    else:
        state.S += dsdt * system.dt
    
    #making sure the population of New York is never exceeded
    state.U += dudt * system.dt
    state.I += didt * system.dt
    state.E += dedt * system.dt
    state.R += drdt * system.dt
    state.D += dddt * system.dt
    state.N += dndt * system.dt
    
    return State(S=state.S, U=state.U, I=state.I, E=state.E, R=state.R, D=state.D, N=state.N)    
```

This function runs the simulation of the model by running update_function repeatedly and storing the stock values at each time step in a frame.

```python
def run_simulation(system, state, update_function, calc_beta, num_init, 
                   num_early_rash, num_rash, immunize):
    """Runs a simulation of the model.
    
    system: System object
    state: State object
    update_function: Function that updates the stocks.
    calc_beta: Function that updates beta.
    num_init: Number of people met during initial symptoms. (Most people contacted)
    num_early_rash: Number of people met during early rash. (Few people contacted)
    num_rash: Number of people met during rash/scabs. (Few people contacted)
    
    Returns a frame."""
    
    unpack(system)
    
    frame = TimeFrame(columns = state.index) 
    frame.row[0] = state
    
    for t in linrange(t0,tend):
        frame.row[t+1] = update_function(frame.row[t], t, system, calc_beta, num_init, 
                                         num_early_rash, num_rash, immunize)
    
    return frame
```

This function plots all of our stocks and labels the graph.

```python
def plot_results(frame, title):
    plot (frame.S, label = "Susceptible")
    plot (frame.U, label = "Incubation")
    plot (frame.I, label = "Initial Symptoms")
    plot (frame.E, label = "Early Rash")
    plot (frame.R, label = "Rash/Scabs")
    plot (frame.D, label = "Dead")
    plot (frame.N, label = "Recovered/Immune")
    decorate(title = title,
            xlabel = "Hours",
            ylabel = "People *10^7")
    #moves legend to the right of the graph so it doesn't overlap
    legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### Numerical Assumptions ###
When we run our simulation, we assume that 1000 people have been infected by smallpox and are in the incubation stage. We also assume that there is one person in the initial symptoms stage, rash stage, and rash/scab stage. CDC stated that the average transmission rate of smallpox is 60%. In addition, they also stated that the early rash stage was the most infectious, followed by the rash/scab stage, and then the initial symptoms stage. While ensuring that the total average of transmission rates is 60%, we tried to reflect the different tranmission rates. 

Our time step for this model is 1 hour (1/24). We go for 4392 time steps which is 6 months.
#### Without Vaccinations or Quarantine ####
We assume that people in the initial symptoms stage would interact with 4 people daily while the people in the early rash and rash/scabs stages would interact with 1 person daily. This is because people would very clearly know they have smallpox in the latter two stages while the people in the initial stages could potentially believe that they have the flu or something less threatening in nature. We also came up with these numbers specifically because the CDC said that a person with smallpox would approximately infect 3 people per day. So, if you multiply each of these values to their corresponding transmission rates, the results is 3 people.
#### With Vaccinations ####
Using the total number of hospitals in New York city (70 hospitals), we estimated that each hospital could successfully vaccinate 1000 people per day due to the influx of patients. This would mean that the city of New York could vaccinate 70,000 people per day or about .81% of their population (this number is extended in the calculations). We assume that people who are vaccinated move from susceptible to recovered/immune. 
#### With Quarantine ####
We assume that people in the early rash and rash/scab stages contact no one and are put in quarantine. We do not assume that people in the initial symptoms stage always make it to quarantine because they might be thought to have a different illness.

## Results ##
### Without Vaccinations or Quarantine ###

```python
state=make_state(8.622*10**6, 1000, 1, 1, 1, 0, 0)
system=make_system(0.4, 0.8, 0.6, (1/24), 0, 4392)
results=run_simulation(system, state, update_function, calc_beta, 4, 1, 1, 0)
plot_results(results,"Smallpox Outbreak In NY City Without Vaccinations or Quarantine")
print ("Recovered/Immune: ")
get_last_value(results.N)
```

### With Vaccinations and Quarantine ###

```python
state=make_state(8.622*10**6, 1000, 1, 1, 1, 0, 0)
system=make_system(0.4, 0.8, 0.6, (1/24), 0, 4392)
results=run_simulation(system, state, update_function, calc_beta, 4, 0, 0, 0.00811782442)
plot_results(results,"Smallpox Outbreak In NY City With Vaccinations and Quarantine")
print ("Recovered/Immune: ")
get_last_value(results.N)
```

In the response where neither vaccines nor quarantine is used, approximately 4,852,842 people survive the outbreak. When quarantine is used, approximately 5,190,245 people survive. So, 337,403 more people survive. When just vaccines are used, approximately 5,712,947 people survive which is 522,702 more people than when just quarantine is used. When vaccines and quarantine are used, 6,195,597 people survive which is 482,650 more peope than when just vaccines are used and 1,342,755 more people than when no response happens. As expected, when both quarantine and vaccines are used the largest number of people survive. There is a bump in the graphs whenever vaccines are used because people are becoming immune every time step without going through all the stages of smallpox. 

## Interpretation ##
Due to lack of data online, we were forced to assign our own transmission rates to each stage of smallpox that averaged to the known overall transmission rate. In addition, the number of people contacted daily in each stage was also calculated based off overall CDC data for how many people someone with smallpox would infect daily. The percent of people that the hospitals of New York City could vaccinate every day may also not be correct because we assumes there would be clinics every day.

The model's question was reached through editing it repeatedly, or iteratively. In addition, the process used to reach the final beta for each time step was revised until the resulting beta made sense. Lastly, the fact that the output of a function changes when state.S = 0 was added later to make sure that there is no dividng by zero.
